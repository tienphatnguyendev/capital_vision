```python
from azure.storage.blob import BlobServiceClient, BlobClient
import os

def download_data_from_blob(connection_string, container_name, download_folder):
    """
    Downloads files from an Azure Blob Storage container to a local directory.

    Parameters:
    - connection_string: Your Azure Blob Storage account connection string.
    - container_name: The name of the Azure Blob Storage container from which files will be downloaded.
    - download_folder: The local directory to which the files will be downloaded.
    """
    # Create the BlobServiceClient object
    blob_service_client = BlobServiceClient.from_connection_string(connection_string)
    # Get the container client
    container_client = blob_service_client.get_container_client(container_name)
    
    # Create local directory if it does not exist
    if not os.path.exists(download_folder):
        os.makedirs(download_folder)
    
    # List all blobs in the container
    blob_list = container_client.list_blobs()
    for blob in blob_list:
        blob_client = container_client.get_blob_client(blob)
        download_file_path = os.path.join(download_folder, blob.name)

        # Create directory structure if it does not exist
        os.makedirs(os.path.dirname(download_file_path), exist_ok=True)
        
        print(f"Downloading blob to {download_file_path}...")
        # Download the blob
        with open(download_file_path, "wb") as download_file:
            download_file.write(blob_client.download_blob().readall())
        print(f"Downloaded {blob.name} successfully.")

# Example usage
connection_string = AZR_CONNECTION_STRING
container_name = "asxcompdata"
download_folder = "FinancialData"

# Call the function to start downloading files
if not os.path.exists(download_folder):
    download_data_from_blob(connection_string, container_name, download_folder)
```

```python
def load_financial_data():
    """
    Loads HTML content for each ASX code from files stored in a structured directory.
    
    Returns:
    - A dictionary where each key is an ASX code and the value is the HTML content of the file.
    """
    base_dir = 'FinancialData'
    financial_data = {}
    SYMBOLS = pd.read_csv('20200501-asx-listed-companies.csv', skiprows=1)['Code']
    for asx_code_dir in os.listdir(base_dir):
        if asx_code_dir in SYMBOLS:
            print(asx_code_dir)
        if (asx_code_dir == '.DS_Store') or (asx_code_dir.lower() not in SYMBOLS.str.lower().values):
            continue
        asx_code_path = os.path.join(base_dir, asx_code_dir)
        for year_dir in os.listdir(asx_code_path):
            year_path = os.path.join(asx_code_path, year_dir)
            raw_html_dir = os.path.join(year_path, 'RawHTML')
            if os.path.isdir(raw_html_dir):
                for file_name in os.listdir(raw_html_dir):
                    if file_name.endswith('.html'):
                        file_path = os.path.join(raw_html_dir, file_name)
                        print(file_path)
                        with open(file_path, 'r', encoding='utf-8') as file:
                            html_content = file.read()
                            asx_code = file_name.split('.')[0]
                            financial_data[asx_code] = html_content
                            print(f'File loaded: {file_path}')
    
    return financial_data
```

```python
comp_data = load_financial_data()
```

```python
def retrieve_dataframes(dfs):
    total_df = 0
    df_names = [
        "Profit & Loss",
        "Balance Sheet",
        "Cash Flow",
        "Per Share Statistics",
        "Sundry Analysis",
        "Growth Rates",
        "Ratio Analysis",
        "Asset Base",
    ]
    res = {}
    for idx, df in enumerate(dfs):
        if (df.shape == (0,1)) and ((df_name := df.columns.values[0]) in df_names):
            if not dfs[idx+2].empty:
                res[df_name.lower().replace(' ', '_').replace('&', 'n')] = dfs[idx+2].rename(columns={dfs[idx+2].columns[0]: 'metrics'})
    return res
```

```python
comp_info = {}
for asx_code, html_content in comp_data.items():
    try:
        tables = pd.read_html(StringIO(html_content))
        comp_info[asx_code] = retrieve_dataframes(tables)
    except ValueError as e:
        if str(e) == "No tables found":
            print(f"No tables found for {asx_code} in the HTML content.")
        else:
            raise
```

```python
def melt_df(df):
    res_df = df.melt(id_vars=['metrics'],
                     value_vars= df.columns.values[1:],
                     var_name='year',
                     value_name='value')
    return res_df
```

```python
def concat_comp_dfs(dfs):
    concat_df = pd.DataFrame()
    for stm_name, statement in dfs.items():
        sheet_df = melt_df(statement)
        sheet_df['statement'] = stm_name
        concat_df = pd.concat([concat_df, sheet_df])
    return concat_df
```

```python
combined_df = pd.DataFrame()
for asx_code in comp_info:
    comp_df = concat_comp_dfs(comp_info[asx_code])
    comp_df['code'] = asx_code
    combined_df = pd.concat([combined_df, comp_df])
    
```

```python
def get_dict_depth(d, depth=0):
    """
    Recursively finds the depth of nesting in a dictionary.

    Args:
    d (dict): The dictionary to check.
    depth (int): The current depth level, defaults to 0.

    Returns:
    int: Maximum depth of nesting found in the dictionary.
    """
    if isinstance(d, dict):
        return max([get_dict_depth(v, depth + 1) for v in d.values()], default=depth)  # Use default to handle empty dicts
    return depth


def store_csv_data(financial_data, SYMBOLS, depth):
    """
    Stores each stock's financial data as a CSV file.
    
    Parameters:
    - financial_data: Dictionary with ASX codes as keys and another dictionary as values,
                      where the inner dictionary has keys as DataFrame names and values as pandas DataFrames.
    - SYMBOLS: Not directly used in this function, may be considered for filtering or validation if needed.
    """
    output_dir = 'CleanedData'
    os.makedirs(output_dir, exist_ok=True)  # Ensure the directory exists

    # Iterate over the financial_data dictionary
    for stock_symbol, dataframes in financial_data.items():
        stock_dir = os.path.join(output_dir, stock_symbol)
        os.makedirs(stock_dir, exist_ok=True)  # Create a directory for each stock symbol if it doesn't exist
        
        if depth == 1:
            csv_path = os.path.join(stock_dir, f'hist_prices_20120501_20230601.csv')
            try:
                dataframes.to_csv(csv_path, index=False)
                print(f'Saved CSV for {stock_symbol} at {csv_path}')
            except Exception as e:
                print(f'Error saving {csv_path}: {str(e)}')
                
                
        if depth == 2:
            # Iterate over each DataFrame associated with the stock symbol
            for df_name, df in dataframes.items():
                csv_path = os.path.join(stock_dir, f'{df_name}.csv')
                try:
                    df.to_csv(csv_path, index=False)
                    print(f'Saved CSV for {stock_symbol} at {csv_path}')
                except Exception as e:
                    print(f'Error saving {stock_symbol}/{df_name}.csv: {str(e)}')
```
