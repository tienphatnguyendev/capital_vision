```python
import os
import pandas as pd

def load_hist_price(directory):
    """
    Loads each stock's financial data from CSV files.
    
    Parameters:
    - directory: The base directory where the CSV files are stored.
    - depth: The depth of the file structure (1 or 2).
    
    Returns:
    - A dictionary with ASX codes as keys and pandas DataFrames (or dictionaries of DataFrames) as values.
    """
    financial_data = {}

    # Traverse the directory containing stock symbols
    for stock_symbol in os.listdir(directory):
        stock_path = os.path.join(directory, stock_symbol)
        if os.path.isdir(stock_path):
            csv_path = os.path.join(stock_path, 'hist_prices_20120501_20230601.csv')
            try:
                financial_data[stock_symbol] = pd.read_csv(csv_path)
                print(f'Loaded CSV for {stock_symbol} from {csv_path}')
            except Exception as e:
                print(f'Error loading {csv_path}: {str(e)}')
    return financial_data

def load_fundamental_metrics(directory):
    """
    Loads financial data from CSV files stored in a structured directory.
    
    Parameters:
    - directory: The path to the directory containing all the stock symbol subdirectories.
    
    Returns:
    - A nested dictionary where the outer keys are stock symbols, and the values are dictionaries
      mapping data types (derived from file names) to pandas DataFrames.
    """
    financial_data = {}
    
    # Traverse the directory containing stock symbols
    for stock_symbol in os.listdir(directory):
        stock_path = os.path.join(directory, stock_symbol)
        if os.path.isdir(stock_path):  # Ensure it's a directory
            data_frames = {}
            # Traverse each file within the stock symbol's directory
            for filename in os.listdir(stock_path):
                if filename.endswith('.csv'):  # Ensure it's a CSV file
                    file_path = os.path.join(stock_path, filename)
                    df_name = filename[:-4]  # Remove '.csv' from filename to use as key
                    try:
                        df = pd.read_csv(file_path)
                        data_frames[df_name] = df
                    except Exception as e:
                        print(f'Failed to load {file_path}: {e}')
            
            # Store the loaded data in the main dictionary
            financial_data[stock_symbol] = data_frames
    
    return financial_data
```

```python
directory = 'CleanedData'
hist_prices, comp_data = load_hist_price(directory),load_fundamental_metrics(directory)
```
